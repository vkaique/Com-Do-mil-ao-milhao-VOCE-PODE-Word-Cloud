## Imports

Para realizar um Word Cloud é necessário importar as libs Pandas, MatPlotLib e WordCloud.

```python
#IMPORTING THE NECESSARY PACKAGES
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
```

## Import ntlk para stopword em português

Dentro da lib wordcloud há uma lista de stopwords que pode ser utilizada em textos na lingua inglesa.

```python
stopwords_en = STOPWORDS
stopwords_en
```

 Contudo, como estamos utilizando textos em português devemos utilizar uma lista contendo os stopwords da língua portuguesa. Para tal, podemos realizar dois procedimentos:

1) Criar uma lista própria

2) Utilizar a lista da lib nltk a qual contém mais de 200 stopwords

```python
import nltk
```

```python
nltk.download('stopwords')
stopwords = nltk.corpus.stopwords.words('portuguese')
```

```python
stopwords[:13]
```

## Do mil ao milhão - livro

Para tal execução utilizamos o livro "Do mil ao milhão sem cortar o cafezinho" de Thiago Nigro em formato .csv o qual está disponível em meu perfil no GitHub.

```python
#IMPORT THE COMPLETE BOOK FILE
DATA_PATH = "https://raw.githubusercontent.com/vkaique/Com-Do-mil-ao-milhao-VOCE-PODE-Word-Cloud/main/domilaomilhao.csv"
book = pd.read_csv(DATA_PATH)
```

```python
book.columns
```

```python
#MISSING VALUES
print("Valores ausentes: ", book.DoMilAoMilhao.isnull().sum())

#ELIMINATING MISSING VALUES
DoMilAoMilhao = book.dropna(subset=['DoMilAoMilhao'], axis=0 )['DoMilAoMilhao']
```

```python
#CONCATENATE THE WORDS
#TOTAL WORDS IN THE BOOK
DoMilAoMilhao = " ".join(s for s in DoMilAoMilhao)
print("Quantidade de Palavras no livro: {}".format(len(DoMilAoMilhao)))
```

```python
#STOPWORD LIST
#stopwords = set(STOPWORDS)

#GENERATE A WORDCLOUD
wordcloud = WordCloud(stopwords=stopwords,
                      background_color="black").generate(DoMilAoMilhao)

#SHOWING THE FINAL IMAGE
fig, ax = plt.subplots(figsize=(10,6))
ax.imshow(wordcloud, interpolation='bilinear')

plt.tight_layout()
```

```python
stopwords = set(STOPWORDS)
stopwords.update(["em", "uma", "um", "para", "de", "não", "que", "são", "ao", 
                  "na", "se", "por", "como", "seu", "da", "mais", "mas", "ou", 
                  "ele", "os", "dos", "sua", "ter", "esse", "isso", "ser", "é", "o", "e"])

wordcloud = WordCloud(stopwords=stopwords,
                      background_color="black",
                      width=1600, height=900).generate(DoMilAoMilhao)

fig, ax = plt.subplots(figsize=(16,8))
ax.imshow(wordcloud, interpolation='bilinear')
ax.set_axis_off()

plt.imshow(wordcloud)
```

```python
import os
from PIL import Image

import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import gaussian_gradient_magnitude
from wordcloud import WordCloud, ImageColorGenerator


#LOAD BOOK
text = DoMilAoMilhao

#LOAD IMAGE COLOR
book_color = np.array(Image.open("pr.png"))
book_color = book_color[::3, ::3]

#CREATE MASK
book_mask = book_color.copy()
book_mask[book_mask.sum(axis=2) == 0] = 255

stopwords = set(STOPWORDS)
stopwords.update(["em", "uma", "um", "para", "de", "não", "que", "são", "ao", 
                  "na", "se", "por", "como", "seu", "da", "mais", "mas", "ou", 
                  "ele", "os", "dos", "sua", "ter", "esse", "isso", "ser", "é", "o", "e"])

edges = np.mean([gaussian_gradient_magnitude(book_color[:, :, i] / 255., 2) 
for i in range(3)], axis=0)
book_mask[edges > .08] = 255

#CREATE WORD CLOUD
wc = WordCloud(stopwords=stopwords, max_words=2000, mask=book_mask, 
               max_font_size=40, random_state=42, relative_scaling=0)

#GENERATE WORD CLOUD
wc.generate(text)
plt.imshow(wc)

#CREATE IMAGE
image_colors = ImageColorGenerator(book_color)
wc.recolor(color_func=image_colors)
plt.figure(figsize=(10, 10))
plt.imshow(wc, interpolation="bilinear")
wc.to_file("book_new.png")


plt.show()
```
